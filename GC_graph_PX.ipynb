**Библиотека Plotly**

```python
import plotly.express as px
import plotly.graph_objs as go
import pandas as pd
```

```python
# Загрузка данных из CSV файла
df = pd.read_csv('MSFT.csv', sep=',')  # df - data frame
df.head(15)  # Выбор всех данных
```

```python
# Назначение переменных данным из CSV файла
a = df['year']  # Год
b = df['earnings']  # Выручка
c = df['revenue']  # Чистая прибыль
d = df['gross_profit']  # Валовая прибыль
```

Линейный график

```python
# Создание линейного графика
fig = go.Figure()  # Создание графика
fig.add_trace(go.Scatter(x=a, y=b, mode='lines+markers', name='Чистая прибыль'))  # Добавление на график линии
fig.add_trace(go.Scatter(x=a, y=c, mode='lines+markers', name='Выручка'))
fig.add_trace(go.Scatter(x=a, y=d, mode='lines+markers', name='Валовая прибыль', line=dict(color='green', width=5,
                                                                                           dash='dot')))
fig.update_layout(title='Финансовые показатели Microsoft Corporation с 2011-2024гг.')  # Название графика
fig.update_layout(yaxis_title='Млрд. $', xaxis_title='Год', title={'font': dict(size=50), 'x': 0.5},
                  legend_orientation='h', legend_font_size=25)  # Название осей и доп. параметры
fig.update_xaxes(gridcolor='black', minor_griddash='dot', titlefont=dict(size=22), tickfont=dict(size=19))  # Параметры оси х
fig.update_yaxes(titlefont=dict(size=22), tickfont=dict(size=19))  # Параметры оси y
fig.update_traces(hoverinfo='all', hovertemplate='Год: %{x}<br>Объем: %{y}')  # Оформление всплывающей подсказки
fig.write_image('linear_PX.png')  # Сохранение графика в файл
fig.show()  # Вызов графика
```

linear_PX.png

Столбчатая диаграмма

```python
# Создание столбчатой диаграммы
fig = go.Figure(data=[
    go.Bar(x=a, y=c, name='Выручка'),
    go.Bar(x=a, y=d, name='Валовая прибыль'),
    go.Bar(x=a, y=b, name='Чистая прибыль'),
])  # Создание графика со столбцами
fig.update_layout(barmode='overlay', title='Финансовые показатели Microsoft Corporation с 2011-2024гг.') # Задание вида и названия
fig.update_layout(yaxis_title='Млрд. $', xaxis_title='Год', title={'font': dict(size=50), 'x': 0.5},
                  legend_orientation='h', legend_font_size=25)  # Название осей и формат легенды
fig.update_xaxes(titlefont=dict(size=22), tickfont=dict(size=19))  # Формат оси x
fig.update_yaxes(titlefont=dict(size=22), tickfont=dict(size=19))  # Формат оси y
fig.update_traces(hoverinfo='all', hovertemplate='Год: %{x}<br>Объем: %{y}')  # Формат ысплывающей подсказки
fig.write_image('bar_PX.png')  # Сохранение графика в файл
fig.show()  # Вызов графика
```

bar_PX.png

Круговая диаграмма

```python
data = [df['earnings'].sum(), df['revenue'].sum(), df['gross_profit'].sum()]  # Подготовка данных
labels = ['Чистая прибыль', 'Выручка', 'Валовая прибыль']  # Наименование данных
fig = px.pie(df, values=data, names=labels, title='Общие финансовые показатели Microsoft Corporation с 2011-2024гг.')  # Создание даграммы
fig.update_layout(title={'font': dict(size=25), 'x': 0.5}, legend_title='Условные обозначения:',
                  legend=dict(font=dict(size=20)), font={'size': 22})  # Формат легенды и подписей
fig.update_traces(textposition='inside', textinfo='percent+label')  # Положение и вид подписей
fig.write_image('circular_PX.png')  # Сохранение графика в файл
fig.show()  # Вызов графика
```
